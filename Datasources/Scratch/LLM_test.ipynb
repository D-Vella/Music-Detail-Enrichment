```python
import sys
import os
sys.path.append(os.path.abspath('../..'))
from Datasources.Wikidata import get_from_wikidata
from Datasources.llm_client import llm_page_entry
```

```python
import wikipediaapi
wiki_wiki = wikipediaapi.Wikipedia(user_agent='MusicDetailEnrichmentBot/1.0 (https://github.com/D-Vella/Music-Detail-Enrichment)', language='en')

main_title, discography_title = get_from_wikidata("Q706748") #powerwolf
main_page = wiki_wiki.page(main_title)
discography_page = wiki_wiki.page(discography_title)
```

```python
summary = main_page.summary 
discography = discography_page.text

llm_data = "Summary:\n" + summary + "\n\n" + "Discography:\n" + discography

result = llm_page_entry(llm_data)   
print(result["overview"])
print(result["sound"])
print(result["activity"])
print(result["discography"])
print(result["sub_genres"])
```
