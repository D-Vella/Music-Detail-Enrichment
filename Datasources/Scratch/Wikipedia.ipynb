```python
import sys
import os
sys.path.append(os.path.abspath('../..'))
from Datasources.Wikidata import get_from_wikidata
```

```python
def print_sections(sections, level=0):
    for s in sections:
        print("%s: %s - %s" % ("*" * (level + 1), s.title, s.text[0:40]))
        print_sections(s.sections, level + 1)
```

```python
import wikipediaapi
wiki_wiki = wikipediaapi.Wikipedia(user_agent='MusicDetailEnrichmentBot/1.0 (https://github.com/D-Vella/Music-Detail-Enrichment)', language='en')

main_title, discography_title = get_from_wikidata("Q706748")
main_page = wiki_wiki.page(main_title)
discography_page = wiki_wiki.page(discography_title)
print(discography_page)

if main_page.exists():
    print("Page - Title: %s" % main_page.title)
    print("Page - Summary: %s" % main_page.summary[0:60])

if discography_page.exists():
    print("Page - Title: %s" % discography_page.title)
    print("Page - Summary: %s" % discography_page.summary[0:60])
```

```python
print_sections(discography_page.sections)
```
