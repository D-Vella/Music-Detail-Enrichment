```python
import musicbrainzngs as mb
mb.set_useragent("Music Detail Enrichment", "0.1", "https://github.com/D-Vella/music-detail-enrichment")
test_artist_name = "Powerwolf"
```

```python
result = mb.search_artists(test_artist_name, limit=5)
for artist in result['artist-list']:
    print(u"{id}: {name}".format(id=artist['id'], name=artist["name"]))
```

```python
import json

artist_id = "3ef77bdc-b1ce-4e44-aa4b-acfa763c5284"
try:
    result = mb.get_artist_by_id(artist_id,includes=["release-groups","url-rels","tags"],release_type=["album"])
except mb.WebServiceError as exc:
    print("Something went wrong with the request: %s" % exc)
else:
    artist = result["artist"]
    tags = artist["tag-list"]
    links = artist["url-relation-list"]

    print(json.dumps(result, indent=4))
```

```python
import pycountry

homepage = [link["target"] for link in links if link["type"] == "official homepage"][0]
country = pycountry.countries.get(alpha_2=artist["country"])
albums = [album["title"] for album in artist["release-group-list"] if album["primary-type"] == "Album" and "secondary-type-list" not in album]

print(tags)
```

```python
print(f"Artist name = {artist['name']}")
print(f"Artist mbid = {artist['id']}")
print(f"Artist homepage = {homepage}")
print(f"Artist country = {country.name}")
print(f"Artist albums = {len(albums)}")
```
