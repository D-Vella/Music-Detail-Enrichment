```python
from notion_client import Client
from notion_client.helpers import collect_paginated_api
from dotenv import load_dotenv
import os
import json
# 1. Load the .env file: 
load_dotenv()

# 2. Access the variables using os.environ
notion_token = os.environ.get("NOTION_TOKEN")
notion_database_id = os.environ.get("NOTION_DATABASE_ID")

notion = Client(auth=notion_token)
```

```python
testing = collect_paginated_api(
    notion.data_sources.query, data_source_id=notion_database_id
)
```

```python
def get_text(prop):
    """Rich text or title field → plain string"""
    items = prop.get("title") or prop.get("rich_text") or []
    return items[0]["plain_text"] if items else None

def get_select(prop):
    """Single select field → string"""
    sel = prop.get("select")
    return sel["name"] if sel else None

def get_multi_select(prop):
    """Multi-select field → list of strings"""
    return [item["name"] for item in prop.get("multi_select", [])]

def get_checkbox(prop):
    """Checkbox field → bool"""
    return prop.get("checkbox", False)

def get_url(prop):
    """URL field → string"""
    return prop.get("url")

def get_date(prop):
    """Date field → start date string (YYYY-MM-DD)"""
    d = prop.get("date")
    return d["start"] if d else None

def get_number(prop):
    """Number field → int or float"""
    return prop.get("number")

def get_relation_ids(prop):
    """Relation field → list of related page IDs"""
    return [rel["id"] for rel in prop.get("relation", [])]


def extract_artist(page):
    props = page["properties"]
    return {
        "id":                   page["id"],
        "artist_name":          get_text(props["Artist Name"]),
        "status":               get_select(props["Status"]),
        "high_level_genre":     get_select(props["High Level Genre"]),
        "subgenre":             get_multi_select(props["Subgenre"]),
        "country":              get_text(props["Country"]),
        "label":                get_text(props["Label"]),
        "rating":               get_select(props["Rating"]),
        "format_owned":         get_multi_select(props["Format Owned"]),
        "website":              get_url(props["Website"]),
        "bandcamp_url":         get_url(props["Bandcamp url"]),
        "following_on_bandcamp":get_checkbox(props["Following on Bandcamp"]),
        "seen_live":            get_checkbox(props["Seen Live"]),
        "mb_id":                get_text(props["MB_ID"]),
        "how_i_found_them":     get_text(props["How I Found Them"]),
        "added":                get_date(props["Added"]),
        "last_listened":        get_date(props["Last Listened"]),
        "notes":                get_text(props["Notes"]),
        "Similar_to":           get_relation_ids(props["Similar to"]),
    }
```

```python
for i, page in enumerate(testing):
    artist = extract_artist(page)
    print(f"Artist {i}: {artist['artist_name']}-{artist['id']}")
```
