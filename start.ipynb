```python
from dotenv import load_dotenv
import os

load_dotenv('.env')
token = os.getenv('API_TOKEN')
database_id = os.getenv('DATABASE_ID')
print(token)
```

```python
import requests
import base64
import json

url = f"https://api.notion.com/v1/databases/{database_id}"
api_ver = '2026-03-11'


headers = {
    "Notion-Version": f"{api_ver}",
    "Authorization": f"Bearer {token}"
}

response = requests.get(url, headers=headers)
datasource_id = response.json()['data_sources'][0]['id']
print(datasource_id)
```

```python
#Get datasource details:
datasource_url = f"https://api.notion.com/v1/data_sources/{datasource_id}"
response = requests.get(datasource_url, headers=headers)
print(json.dumps(response.json(), indent=4))
```

```python
#Query datasource:
datasource_url = f"https://api.notion.com/v1/data_sources/{datasource_id}/query"
fields_list = ['title','Country','Label','Website','Genre']
params = {
    'filter_properties[]': ['title', 'Country', 'Label', 'Website', 'Genre']
}
payload = {
    "sorts": [
        {
            "property": "title",
            "direction": "ascending"
        }
    ],
    "page_size": 10,
    "in_trash": False,
    "result_type": "page"
}

response = requests.post(datasource_url, headers=headers, json=payload, params=params)
print(json.dumps(response.json(), indent=4))
```
